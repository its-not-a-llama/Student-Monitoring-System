```python
"""
menu.ipynb
Student Monitoring System

GUI for Student Monitoring System using IpyWidgets.
This system allows datapreprocesing, running student queries and running an
analysis for performance
"""

import os
import ipywidgets as widgets
from IPython.display import display, clear_output
import CWpreprocessing
import underperformingStudent
import studentPerformance
import testResults
import warnings
warnings.filterwarnings('ignore')  # Suppress all warnings

# Create widgets
student_id = widgets.IntText(description='Student ID:') # asks for student id

test_dropdown = widgets.Dropdown(
    options=['All', 'Mock Test', 'Formative Test 1', 'Formative Test 2', 
             'Formative Test 3', 'Formative Test 4', 'Summative Test'],
    description='Test:' # drop down menu for test selection
)
btn_preprocess = widgets.Button(
    description='Preprocess Data', button_style='primary') 

btn_submit_2 = widgets.Button(description='Submit', button_style='success')

btn_underperf = widgets.Button(description= 'Underperformance')
btn_underperf.style.button_color = 'darkred'
btn_underperf.style.text_color = 'white'

btn_clear = widgets.Button(description = 'Clear', button_style = 'warning')
output = widgets.Output()

# Functions (add doc strings)
def on_preprocess(b):
    """Run preprocessing file and creates database containing test tables"""
    with output:
        clear_output()
        CWpreprocessing.main()
        print('Preprocessing Complete!')


def on_submit_2(b):
    """ 
    Runs main function of testResults or studentPerformance depending on 
    test_dropdown value to allow student query
    """
    with output:
        clear_output()
        sid = student_id.value
        test_id = test_dropdown.value

        if test_dropdown.value == 'All':
            testResults.main(sid)
            return
        
        elif test_dropdown.value == 'Mock Test':
            test_id = 'm'
        elif test_dropdown.value == 'Formative Test 1':
            test_id = 'f1'
        elif test_dropdown.value == 'Formative Test 2':
            test_id = 'f2'
        elif test_dropdown.value == 'Formative Test 3':
            test_id = 'f3'
        elif test_dropdown.value == 'Formative Test 4':
            test_id = 'f4'
        elif test_dropdown.value == 'Summative Test':
            test_id = 's'


        studentPerformance.main(test_id, sid)
        

def on_clear(b):
    """ Function clears output"""
    with output:
        clear_output()
        #print()

def on_underperform(b):
    """ Runs main function from underperformingStudent to run analysis"""
    with output:
        clear_output()
        underperformingStudent.main()
  

# Connect buttons to functions
btn_preprocess.on_click(on_preprocess)
btn_underperf.on_click(on_underperform)
btn_submit_2.on_click(on_submit_2)
btn_clear.on_click(on_clear)

# Display
menu = widgets.VBox([
    widgets.HTML(
        '<h1 style = "text-align: center" >📚Student Monitoring System📚</h1>'),
    widgets.HBox([btn_preprocess, btn_underperf], 
                 layout = widgets.Layout(justify_content = 'center')),
    widgets.HTML(
        '<h3 style = "text-align: center" > Student Query </h3>'),
    widgets.VBox([student_id, test_dropdown], layout=widgets.Layout(align_items='center')),
    widgets.HBox([btn_clear, btn_submit_2], layout=widgets.Layout(justify_content='center')),
    ])

display(menu)
display(output)
```
